```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
sigma = 10
r = 28
b = 8/3
def func(t,y):
    return [-sigma*y[0] + sigma*y[1], 
            (r-y[2])*y[0] - y[1],
            y[0]*y[1] - b*y[2]]
y0 = [0, 1, 0] 
t_span = [0,50]
sol = solve_ivp(func,t_span,y0)
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')

ax.plot(sol.y[0], sol.y[1], sol.y[2])
ax.set_title("Attractor Lorenza")
fig.set_size_inches(18.5, 10.5)
plt.show()
```
